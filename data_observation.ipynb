```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
synthetic_data = pd.read_csv("data/synthetic_data.csv")
feature_df = pd.read_csv("data/preprocessed_df.csv")
balanced_synthetic_data = pd.read_csv("data/balanced_synthetic_data.csv")
balanced_feaure_df = pd.read_csv("data/balanced_feature_df.csv")
```

### Create pie chart observate the label distribution of each dataset

```python
#Count label
label_counts = synthetic_data['Label'].value_counts()

#create porportion pie chart
plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.axis('equal')  
plt.title('Delivery Status (On-time vs Late)')
plt.show()
```

```python
label_counts = balanced_synthetic_data['Label'].value_counts()

plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.axis('equal')  
plt.title('Delivery Status (On-time vs Late) for Synthetic Data')
plt.show()
```

```python
label_counts = balanced_feaure_df['Label'].value_counts()

plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.axis('equal')  
plt.title('Delivery Status (On-time vs Late) for Real Data')
plt.show()
```

```python
label_counts = feature_df['Label'].value_counts()

plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.axis('equal')  
plt.title('Delivery Status (On-time vs Late)')
plt.show()
```

### correlation matrix of features and label in synthetic data and real data

```python
real_df = pd.read_excel("table_with_borders.xlsx")
```

```python
#select columns from real dataset that have numerical data types (int, float)
feature = feature_df.select_dtypes(include=[np.number])
corr_matrix = feature.corr()

#correlation matrix ploting
plt.figure(figsize=(14,14))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()
```

```python
#same for synthetic data
data = synthetic_data.select_dtypes(include=[np.number])
corr_matrix = data.corr()

plt.figure(figsize=(14, 14))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()
```
